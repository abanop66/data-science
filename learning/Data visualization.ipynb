```python
import pandas 
import numpy 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.DataFrame({"x":[1,2,3,4,5] , "y":[6,7,8,9,10]})
```

```python
#remember hist 
df.hist(bins=5)
```

```python
plt.plot(df)
```

```python
plt.bar(df["x"],df["y"] ,label="df1")
plt.title("aba")
plt.legend()
```

```python
plt.plot([],[], label="task1",color="black",linewidth=5)
plt.plot([],[], label="task2",color="w",linewidth=5)
plt.plot([],[], label="task3",color="y",linewidth=5)
plt.plot([],[], label="task4",color="r",linewidth=5)
plt.plot([],[], label="task4",color="r",linewidth=5)

plt.pie(df["x"], colors=["black","g","y","r","b"], startangle=1) 
plt.legend()
```

```python
#if you want specifc figure .... plt.figure

plt.figure(figsize=(8,4))

#if i have x and y1 , y2 , y3 ... we need stackplot

emp=[1,2,3,4]
task1=[20,30,50,70]
task2=[20,40,50,70]
task3=[20,70,50,40]
task4=[70,30,50,10]

#why empty plots ...... to use the legend
plt.plot([],[], label="task1",color="black",linewidth=5)
plt.plot([],[], label="task2",color="w",linewidth=5)
plt.plot([],[], label="task3",color="y",linewidth=5)
plt.plot([],[], label="task4",color="r",linewidth=5)

plt.stackplot(emp,task1,task2,task3,task4,colors=["black","w","y","r"])
plt.title("performance")
plt.legend()
plt.xlabel("employes")
plt.ylabel("tasks")
```
